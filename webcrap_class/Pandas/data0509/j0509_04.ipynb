### dropna - Nan처리:지우기

## axis : index,column

```python
df['학교']='디지털고'
import numpy as np
df['학교']=np.nan
df
```

```python
#Nan이 들어가 있는 컬럼을 전체 삭제
df.dropna(axis='columns',inplace=True)
```

```python
#Nan이 들어가 있는 컬럼에서 Nand 있는것만 삭제
df.dropna(axis='columns',how='all', inplace=True)
```

```python
# Nan 들어가 있는 컬럼에서 1개라도 Nan이 포함되어있으면 삭제
df.dropna(axis='columns',how='any',inplace=True)
```
