```python
import pandas as pd
df=pd.read_csv('../data/stu_score1.csv',index_col='no')
df
```

### id에 i가 들어가있는 학생을 검색하시요

```python
###id에 i가 들어가있는 학생을 검색하시요
filt=df['id'].str.contains('i',na=False)
df.loc[filt,'id']
```

### id에 all이 포함되어있으면서 평균이 67점 이상인 학생을 검색해라


```python
###id에 all이 포함되어있으면서 평균이 67점 이상인 학생을 검색해라
filt=((df['id'].str.contains('all',na=False)),(df['avg']>=67))
df[filt,'id']
```
