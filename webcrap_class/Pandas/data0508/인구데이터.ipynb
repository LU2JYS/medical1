```python
df=pd.read_excel('20230103_인구.xlsx',skiprows=3,index_col='행정기관',usecols='B,E:Y') 
df.head()
```

```python
### rows데이터 불러오기
df[0:1] # 하나부터둘까지 들오고기
df.iloc[0]
df.head(3)
df.tail(3)
df[df.columns[-1]]

df.iloc[0] #0번째 rows데이터 가져오기
df.iloc[0].str #str타입으로 변경
df.iloc[0].str.replace(",","") #쉼표 제거 #replace함수 사용해서 ,를 제거 
```
