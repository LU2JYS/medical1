#### 함수를 사용해서
#### 국어 점수를 20점 더하고 100점 이상의 국어점수를 100점으로 변경하시요

```python
def add_score(kor):
    kor=kor+20
    if kor>100:kor=100
    return kor
```

```python
#함수를 사용해서
# 국어 점수를 20점 더하고 100점 이상의 국어점수를 100점으로 변경하시요

df['국어']=df['국어'].apply(add_score)
df
```

```python
#100점 이상의 국어점수를 100점으로 변경하시요
df['국어']>100
df.loc[df['국어']>100,'국어']='100'
```

```python
import pandas as pd
df = pd.read_excel('../data/score.xlsx',index_col='지원번호')
df
```

```python
# (국+영+수+과+사) >> 총합,평균의 컬럼을 추가하고
# 평점이라는 컬럼을 만들어서 90점이상이면 a, 80이상이면 b ,c,d, 60점미만은 f함수를 사용해서 처리하시요
```

```python
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균']=df['총합']/5
df['평점']=0
df.fillna(0,inplace=True)
df
```

## 함수로 평점 넣기

```python
def grade(score):
    if score >=90 : return "A"
    elif score >=80 : return "B"
    elif score >=70 : return "C"
    elif score >=690 : return "D"
    else : return "F"
```

```python
df['결과']=df['평균'].apply(grade)
```

```python
def result(score):
    txt =''
    if score >= 70 : txt='합격'
    else : txt= "불합격"
    return txt
```

### lamda 함수 사용-------------------------------------------------------

```python
#결과 컬럼,평균 70점 이상 인 학생은 합격 ,불합격

df['결과']=df['평균'].apply(lambda score:'합격' if score>=70 else "불합격")
```

```python
### 결과2 컬럼을 생성 65점 초과 합격 65점은 재시험,65점은 미만은 불합격
df['결과2']=df['평균'].apply(lambda score:'합격' if score>65 else ("재시험" if score==65 else "불합격"))
```

```python
df
```

```python
# 국어+영어 =합쳐서 100점이 넘지않으면  국어점수에 +20 추가해서 출력해라
filt =df['국어']+df['영어']
df.loc[filt<100,'국어']=df.loc[filt<100,'국어']+20
df
```
