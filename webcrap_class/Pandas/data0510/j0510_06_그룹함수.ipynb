```python
# 학교,성별로 국어 영어 수학점수 평균을 출력하시요
df.groupby(['school','gender'])[['kor','eng','math']].mean()
# df.groupby(['school','gender'])[['kor','eng','math']].mean().loc['디지털고'] #디지털고 출력
# df.groupby(['school','gender'])[['kor','eng','math']].mean().sort_values('kor',ascending=False) #높은순 출력
```

```python
df['total'] = df['kor']+df['eng']+df['math']
```

```python
df
```

```python
df['avg']=df['total']/3
```

#### agg(['mean','min','max','sum'])는 한번에 평균 최소 최대 합계를 볼수있당...

```python
# 학교의 학년별로 한번에 평균 최소 최대 합계를 볼수있당...
df.groupby(['school','grade'])['total'].agg(['mean','min','max','sum'])
```
