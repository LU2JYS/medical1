```python
import pandas as pd
df = pd.read_csv('https://bit.ly/fish_csv')
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
# 1. 데이터 전처리 -data,target 분리

fish_target = df['Species'].to_numpy()
fish_input = df.iloc[:,1:].to_numpy()

# train,test세트 분리
train_input,test_input,train_target,test_target = train_test_split(
    fish_input,fish_target, random_state=42
)

# 표준화 작업
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)


```

```python
# 2. 데이터 실습시키기
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(train_scaled,train_target)
```

```python
# 3. 데이터 예측
predict=lr.predict(test_scaled[:5])
print(predict)

# 기울기,y절편의 값을 적용시켜서 z점수 출력
# softmax함수를 적용해서 가장 높은 확률을 채택해서 결정함..
```

```python
# 4. 데이터 정답률 - train,test
print(lr.score(train_scaled,train_target))
print(lr.score(test_scaled,test_target))
```

### 결과값 도출 형태

```python
proba = lr.predict_proba(train_scaled[:5])
np.round(proba,decimals=3)
```

```python
decision = lr.decision_function(test_scaled[:5])
print(np.round(decision,decimals=2))
```

```python

from scipy.special import softmax

proba = softmax(decision,axis = 1)
print(np.round(proba,decimals=3))
```

#### C=30, max_iter=1000 횟수를 늘려 정답률을 높이는 방법이다


```python
# 2. 데이터 실습시키기
from sklearn.linear_model import LogisticRegression

# max_iter : 반복횟수=100 기본설정
# C규제 : 올라가면 규제가 약해짐, 낮아지면 규제가 강해짐, 선형회귀 규제 = alpha값으로 지정
lr = LogisticRegression(C=30, max_iter=1000)
lr.fit(train_scaled,train_target)
```

```python
print(lr.score(train_scaled,train_target))
print(lr.score(test_scaled,test_target))
```
